```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
```

```python
sns.set_style('whitegrid', {'grid.color': '0.7'})
```

## Training

```python
c1_df = pd.read_csv('client_6666.csv')
c1_df.drop(columns='Unnamed: 0', inplace=True)
c1_df
```

```python
c2_df = pd.read_csv('client_6699.csv')
c2_df.drop(columns='Unnamed: 0', inplace=True)
c2_df
```

```python
plt.figure(figsize=(7, 5))

sns.lineplot(data=c1_df, x='epoch', y='train_loss', color='purple', label='Client 1')
sns.lineplot(data=c2_df, x='epoch', y='train_loss', color='green', label='Client 2')
sns.lineplot(x=c1_df['epoch'], y=((c1_df['train_loss'] + c2_df['train_loss']) / 2), color='blue', label='Network Average')

plt.xlabel('Epochs')
plt.ylabel('Average Training Loss')
plt.title('Learning Curves')

plt.savefig('learning_curves.pdf', dpi=1000)
_ = plt.show()
```

```python
plt.figure(figsize=(7, 5))

sns.lineplot(data=c1_df, x='epoch', y='val_loss', color='purple', label='Client 1')
sns.lineplot(data=c2_df, x='epoch', y='val_loss', color='green', label='Client 2')
sns.lineplot(x=c1_df['epoch'], y=((c1_df['val_loss'] + c2_df['val_loss']) / 2), color='blue', label='Network Average')

plt.xlabel('Epochs')
plt.ylabel('Average Validation Loss')
plt.title('Validation Curves')

plt.savefig('validation_curves.pdf', dpi=1000)
_ = plt.show()
```

```python
plt.figure(figsize=(7, 5))

sns.lineplot(data=c1_df, x='epoch', y=c1_df['val_acc'] * 100, color='purple', label='Client 1')
sns.lineplot(data=c2_df, x='epoch', y=c2_df['val_acc'] * 100, color='green', label='Client 2')
sns.lineplot(x=c1_df['epoch'], y=((c1_df['val_acc'] + c2_df['val_acc']) / 2) * 100, color='blue', label='Network Average')

plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy')

plt.savefig('val_accs.pdf', dpi=1000)
_ = plt.show()
```

## Distributions

```python
general_path = f'Implementation/mislabeled_subsets/'
flip_percentages = [0.05, 0.1, 0.15]
subsets = [0, 1, 2, 3]
```

```python
fig, axes = plt.subplots(len(flip_percentages), len(subsets) + 1, figsize=(20, 14))

for i, percent in enumerate(flip_percentages):
    combined_labels = []

    # Plot the total distribution for each flip percentage
    axes[i, 0].set_title(f'Total Distribution - {percent} flipped')

    for j, subset in enumerate(subsets):
        with open(general_path + f'{percent}/subset_{subset}.pkl', 'rb') as file:
            sub = pickle.load(file)
            labels = [label for _, label in sub]
            combined_labels.extend(labels)

        # Plot individual histograms for each subset
        axes[i, j + 1].hist(labels, bins=range(11), align='left', rwidth=0.8, color='blue', edgecolor='black')
        axes[i, j + 1].set_xticks(range(10))
        axes[i, j + 1].set_xlabel('Classes')
        axes[i, j + 1].set_ylabel('Data Samples')
        axes[i, j + 1].set_title(f'Subset {subset} - {percent} flipped')

    # Plot the total distribution for each flip percentage
    axes[i, 0].hist(combined_labels, bins=range(11), align='left', rwidth=0.8, color='green', edgecolor='black')
    axes[i, 0].set_xticks(range(10))
    axes[i, 0].set_xlabel('Classes')
    axes[i, 0].set_ylabel('Data Samples')

# Adjust layout to prevent overlapping titles
plt.suptitle('IID Experiment: Data Distributions')
plt.tight_layout()
plt.savefig('flipped_distributions.pdf', dpi=1000)
plt.show()
```
