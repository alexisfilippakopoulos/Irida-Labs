```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
c1_df = pd.read_csv('client_6666.csv')
c1_df.drop(columns='Unnamed: 0', inplace=True)
c1_df
```

```python
c2_df = pd.read_csv('client_6699.csv')
c2_df.drop(columns='Unnamed: 0', inplace=True)
c2_df
```

```python
sns.set_style('whitegrid', {'grid.color': '0.7'})
```

```python
plt.figure(figsize=(7, 5))

sns.lineplot(data=c1_df, x='epoch', y='train_loss', color='purple', label='Client 1')
sns.lineplot(data=c2_df, x='epoch', y='train_loss', color='green', label='Client 2')
sns.lineplot(x=c1_df['epoch'], y=((c1_df['train_loss'] + c2_df['train_loss']) / 2), color='blue', label='Network Average')

plt.xlabel('Epochs')
plt.ylabel('Average Training Loss')
plt.title('Learning Curves')

plt.savefig('learning_curves.pdf', dpi=1000)
_ = plt.show()
```

```python
plt.figure(figsize=(7, 5))

sns.lineplot(data=c1_df, x='epoch', y='val_loss', color='purple', label='Client 1')
sns.lineplot(data=c2_df, x='epoch', y='val_loss', color='green', label='Client 2')
sns.lineplot(x=c1_df['epoch'], y=((c1_df['val_loss'] + c2_df['val_loss']) / 2), color='blue', label='Network Average')

plt.xlabel('Epochs')
plt.ylabel('Average Validation Loss')
plt.title('Validation Curves')

plt.savefig('validation_curves.pdf', dpi=1000)
_ = plt.show()
```

```python
plt.figure(figsize=(7, 5))

sns.lineplot(data=c1_df, x='epoch', y=c1_df['val_acc'] * 100, color='purple', label='Client 1')
sns.lineplot(data=c2_df, x='epoch', y=c2_df['val_acc'] * 100, color='green', label='Client 2')
sns.lineplot(x=c1_df['epoch'], y=((c1_df['val_acc'] + c2_df['val_acc']) / 2) * 100, color='blue', label='Network Average')

plt.xlabel('Epochs')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracy')

plt.savefig('val_accs.pdf', dpi=1000)
_ = plt.show()
```
